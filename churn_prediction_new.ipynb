Changing Categorical columns into Numerical Columns

```python
# from sklearn.preprocessing import LabelEncoder

# cat_cols = syn_data[['Gender', 'Education_Level', 'Marital_Status', 'Income_Category','Card_Category']]

# # Get the mapping values
# le = LabelEncoder()

# # store the mappings in a dictionary
# mapping = {}

# # transform the columns and store the mapping
# for col in cat_cols:
#     syn_data[col] = le.fit_transform(syn_data[col])
#     mapping[col] = dict(zip(le.classes_, le.transform(le.classes_)))

# # print the mapping
# for col, values in mapping.items():
#     print(f"Mapping for column {col}:")
#     for key, value in values.items():
#         print(f"{key} is mapped to {value}")
#     print("\n")

import pandas as pd
from sklearn.preprocessing import LabelEncoder


# Get the categorical columns
cat_cols = syn_data[['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']]

# Create a dictionary to store the mappings
mapping = {}

# Define a function to apply LabelEncoder to a column
def encode_column(col):
    le = LabelEncoder()
    encoded_col = le.fit_transform(col)
    mapping[col.name] = dict(zip(le.classes_, le.transform(le.classes_)))
    return encoded_col

# Apply LabelEncoder using apply
syn_data[cat_cols.columns] = cat_cols.apply(lambda x: encode_column(x))
```

## Feature Selection

```python
features = syn_data[['Customer_Age', 'Gender','Dependent_count','Education_Level','Marital_Status', 'Income_Category', 'Card_Category', 'Months_on_book','Total_Relationship_Count','Months_Inactive_12_mon','Contacts_Count_12_mon','Credit_Limit']]
target = syn_data['Attrition_Flag']
```

```python
acc = []
model = []
```

Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size = 0.2,random_state =2)
```

## Applying Algorithms

### Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import metrics

DecisionTree = DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5)

DecisionTree.fit(Xtrain,Ytrain)

predicted_values = DecisionTree.predict(Xtest)
x = metrics.accuracy_score(Ytest, predicted_values)
acc.append(x)
model.append('Decision Tree')
print("DecisionTrees's Accuracy is: ", x*100)

print(classification_report(Ytest,predicted_values))
```

### Cross Validation Score for Decision Tree

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(DecisionTree, features, target,cv=5)
score
```

### Saving Decision Tree model

```python
import pickle

DT_pkl_filename = 'models/DecisionTree.pkl'
DT_Model_pkl = open(DT_pkl_filename, 'wb')
pickle.dump(DecisionTree, DT_Model_pkl)
DT_Model_pkl.close()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=20, random_state=0)
RF.fit(Xtrain,Ytrain)

predicted_values = RF.predict(Xtest)

x = metrics.accuracy_score(Ytest, predicted_values)
acc.append(x)
model.append('RF')
print("RF's Accuracy is: ", x)

print(classification_report(Ytest,predicted_values))
```

### Cross Validation Score for Random Forest

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(RF, features, target,cv=5)
score
```

## Confusion Matrix

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score


# Create an ensemble of the four algorithms
ensemble = VotingClassifier(estimators=[
        ('DT', DecisionTree), 
        ('RF', RF), 
        ('LR', LogReg), 
        ('KN', KN)], 
        voting='hard')


ensemble.fit(Xtrain, Ytrain)
y_pred_ensemble = ensemble.predict(Xtest)

# Calculate the confusion matrix and evaluation metrics for the ensemble
cm_ensemble = confusion_matrix(Ytest, y_pred_ensemble, labels=['Attrited Customer', 'Existing Customer'])
acc_ensemble = accuracy_score(Ytest, y_pred_ensemble)
prec_ensemble = precision_score(Ytest, y_pred_ensemble, pos_label='Existing Customer')
rec_ensemble = recall_score(Ytest, y_pred_ensemble, pos_label='Existing Customer')
f1_ensemble = f1_score(Ytest, y_pred_ensemble, pos_label='Existing Customer')


print("Ensemble\n", cm_ensemble)
print("Accuracy:", acc_ensemble)
print("Precision:", prec_ensemble)
print("Recall:", rec_ensemble)
print("F1 score:", f1_ensemble)
```
